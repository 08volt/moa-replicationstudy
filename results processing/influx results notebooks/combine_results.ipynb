```python
algorithms = ["RebalanceStream","CSMOTE","improved_UOB","improved_OOB","WEOB1",
              "WEOB2","ESOS_ELM","UOB","OOB"]
cnt = 0
for drift_type, dict_ds in result_dict.items():
    removeP = 0
    if "P" == drift_type[0]:
        removeP = 1
        
    drift_dir = "Combined/"+drift_type[removeP:]
    if not os.path.exists(drift_dir):
        os.mkdir(drift_dir)
        
    for ds, algs in dict_ds.items():
        file_path = drift_dir + '/' + ds + '.csv'
        with open(file_path, mode='w') as ds_file:
            writer = csv.writer(ds_file, quoting=csv.QUOTE_MINIMAL)
            writer.writerow(['ds', 'alg', 'avg_mem'])
        
        for long_alg, value in algs.items():
            short_alg = ""
            for a in algorithms:
                if a in long_alg:
                    short_alg = a
                    break

            assert short_alg != ""
        
            with open(file_path, mode='a') as ds_file:
                writer = csv.writer(ds_file, quoting=csv.QUOTE_MINIMAL)
                writer.writerow([drift_type[removeP:]+'_'+ds, short_alg, value])
                cnt += 1
print(f"scritti {cnt} valori")
                
            
        
        
            
    
```

```python
def read_file(file_path):
    file = open(file_path)
    reader = csv.reader(file)
    lines= len(list(reader))
    print(f"{file_path} has {lines} lines")
    



for dir_entry in os.listdir("./Combined"):
    if dir_entry == ".DS_Store":
        continue
    print(dir_entry)
    if os.path.isfile("./Combined/" +dir_entry):
        read_file("./Combined/" +dir_entry)
    else:
        for dir_entry1 in os.listdir("./Combined/" +dir_entry ):
            read_file("./Combined/" +dir_entry +"/"+dir_entry1)
            

```
